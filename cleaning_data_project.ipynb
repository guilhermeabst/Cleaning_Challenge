```python
# Importing Pandas 
```

```python
import pandas as pd
```

```python
# Reading CSV File
```

```python
data = pd.read_csv("data_cleaning_challenge.csv")
data.head(10)
```

```python
# Deleting Unuseful Columns
```

```python
data_drop = data.drop(columns=["Unnamed: 9","Unnamed: 10"])
data_drop.head()
```

```python
#Deleting NaN in Row Type Columns
```

```python
data_nona = data_drop[data_drop["Row Type"].notna()]
data_nona.head()
```

```python
# Creating a iteration column to merge the datasets
```

```python
column_iter = []
count = 0 
for i in data_nona["Row Type"]:
    if "first name" in i:
        count = count + 1
    column_iter.append(count)
column_iter
```

```python
data_iter = data_nona
data_iter["Iteration"] = column_iter
data_iter.head()
```

```python
# Deleting rows with wrong values
```

```python
data_no_rowtype = data_iter[data_iter["Row Type"] != "Row Type"]
data_no_rowtype.head()
```

```python
# Creating the new dataset to merge
```

```python
new_columns_data = data_no_rowtype[data_no_rowtype["Row Type"].str.contains("first name")]
new_columns_data.drop(columns=["Speed1","Speed2","Electricity","Effort","Torque","Weight"],inplace= True)
new_columns_data.rename(columns={"Row Type":"First Name","Iter Number":"Last Name","Power1":"Date"}, inplace=True)
new_columns_data.head()
```

```python
# Cleaning the new Dataset
```

```python
new_columns_data["First Name"] = new_columns_data["First Name"].apply(lambda x:x.split(":")[1])
new_columns_data["Last Name"] = new_columns_data["Last Name"].apply(lambda x:x.split(":")[1])
new_columns_data["Date"] = new_columns_data["Date"].apply(lambda x:x.split(":")[1])
new_columns_data.head()
```

```python
# Cleaning the "first name: Person" values from the first data set
```

```python
data_no_firstname = data_no_rowtype[~data_no_rowtype["Row Type"].str.contains("first name")]
data_no_firstname.head()
```

```python
# Merging Datasets and finishing the project
```

```python
complete_data = pd.merge(data_no_firstname,new_columns_data,on="Iteration")
complete_data.head(10)
```
